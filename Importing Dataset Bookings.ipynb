```python
#Translate Column Names
new_names = bookings.columns
new_names=[str.lower(x) for x in new_names]
new_names=['status', 'supplier', 'booking_id', 'roundtrip_id', 'name', 'pickup_date', 
           'pickup_time', 'number_of_passenger', 'vehicle_type', 'route_type', 'route', 
           'total_price', 'total_comission', 'total_driver', 'special_notes', 'city_of_departure', 
           'address', 'hotel', 'flight_number', 'last_update_date', 'booking_date']
bookings.columns=new_names
```

```python
#Convert to Datetimes
#Example of what the source has: 2016/09/01 19:00:00
bookings['pickup_date'] = pd.to_datetime(bookings['pickup_date'])
bookings['last_update_date'] = pd.to_datetime(bookings['last_update_date'])
bookings['booking_date'] = pd.to_datetime(bookings['booking_date'])
```

```python
#Remove some personal bookings that will have strange unconventional IDs.
#bookings.sort_values(len(str(bookings['booking_id'])), ascending=False)
#pd.set_option('display.max_rows', 1000)
filter=bookings['booking_id'].astype(str).map(len)>13
sum(filter) #total 18
bookings=bookings[~filter]
bookings.shape
```

```python
#Remove some cancelled bookings that don't have ID
filter=bookings['booking_id'].isna()
print(sum(filter))
bookings=bookings[~filter]
```

```python
#Convert Number of Passengers to Integer
bookings['number_of_passenger']=bookings['number_of_passenger'].fillna(0)
bookings['number_of_passenger']=bookings['number_of_passenger'].astype(int)
```

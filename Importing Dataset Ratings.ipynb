```python
#Decide which columns are being kept
print(ratings.columns)
all=['id_traslado', 'nombre', 'email', 'texto', 'idioma', 'puntuacion',
       'fecha_traslado', 'fecha_opinion', 'numero_personas', 'vehiculo',
       'origen', 'destino', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14',
       'Unnamed: 15', 'Unnamed: 16']
keeping=['id_traslado', 'texto', 'idioma', 'puntuacion',
       'fecha_traslado', 'fecha_opinion', 'numero_personas', 'vehiculo',
       'origen', 'destino']
ratings=ratings[keeping]
```

```python
#Translate Column Names
ratings.columns
new_names=['booking_id', 'rating_text', 'country', 'rating_score', 'pickup_date',
          'rating_date', 'number_of_passenger', 'vehicle', 'booking_departure', 'booking_destination']
new_names=[str.lower(x) for x in new_names]
ratings.columns=new_names
```

```python
#Convert to Datetimes
#Example of what the source has: 2016/09/01 19:00:00
ratings['pickup_date'] = pd.to_datetime(ratings['pickup_date'])
ratings['rating_date'] = pd.to_datetime(ratings['rating_date'])
```

```python
#Check for rows that don't have ID
filter=ratings['booking_id'].isna()
print(sum(filter))
#ratings=ratings[~filter]
ratings['booking_id']
```

```python
#Checking Values in some Columns.
#booking_departure
#Putting the values in a SET
unique_values_set = set(ratings['booking_departure'])
#Count unique
ratings.booking_departure.nunique()
print(unique_values_set)

#booking_destination
#Putting the values in a SET
unique_values_set = set(ratings['booking_destination'])
#Count unique
ratings.booking_destination.nunique()
print(unique_values_set)

#vehicle
#Putting the values in a SET
unique_values_set = set(ratings['vehicle'])
#Count unique
ratings.vehicle.nunique()
print(unique_values_set)

#rating_score
#Putting the values in a SET
unique_values_set = set(ratings['rating_score'])
#Count unique
ratings.rating_score.nunique()
print(unique_values_set)

#country
unique_values_set = set(ratings['country'])
#Count unique
ratings.country.nunique()
print(unique_values_set)

```

```python
#Scores are not supossed to be 0, they range from 1 to 5. Let's see
filter=ratings['rating_score']==0
ratings=ratings[~filter]
ratings.shape
#There are only 2 that don't have score. Let's not include them.
```

```python
#Convert Number of Passengers to Integer
ratings['number_of_passenger']=ratings['number_of_passenger'].fillna(0)
ratings['number_of_passenger']=ratings['number_of_passenger'].astype(int)
```

```python
#After all the changes, we proceed to save the changes in a file.
import os
outname = 'all_ratings2.0.csv'
outdir = './Datasets'
if not os.path.exists(outdir):
    os.mkdir(outdir)
fullname = os.path.join(outdir, outname)    

ratings.to_csv(fullname, header=True, index=False)
print("Saved!")
```

```python
import folium
from folium import plugins
from folium.plugins import HeatMap
from scipy import spatial
spatial.ConvexHull


#The latitude of Manhattan, New York City, NY, USA is 40.776676, and the longitude is -73.971321
map_hooray = folium.Map(location=[40.776676, -73.971321], tiles='OpenStreetMap', zoom_start = 12) #Cartodb Positron #OpenStreetMap #Cartodb dark_matter

# Ensure you're handing it floats

rba=rb[~((rb['address_lat'].isna())|(rb['address_loc'].isna()))]

rba['address_lat'] = rba['address_lat'].astype(float)
rba['address_loc'] = rba['address_loc'].astype(float)

# Filter the DF for rows, then columns, then remove NaNs
rba = rba[['address_lat', 'address_loc']]
rba = rba.dropna(axis=0, subset=['address_lat','address_loc'])

# List comprehension to make out list of lists
rba = [[row['address_lat'],row['address_loc']] for index, row in rba.iterrows()]

# Plot it on the map
HeatMap(rba, min_opacity=0.22).add_to(map_hooray)

#Add the airports

icon1_image = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcT1TsXGfUKMDRs8FjOgGswjvgAcKVCatsTGng&usqp=CAU'

icon1 = folium.CustomIcon(
    icon1_image,
    icon_size=(90, 45),
    icon_anchor=(22, 94),
)

icon2_image = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTrJgZxChFHlipdAWtJCs_ABgUg5bnaVCKnzw&usqp=CAU'

icon2 = folium.CustomIcon(
    icon2_image,
    icon_size=(90, 45),
    icon_anchor=(22, 94),
)

icon3_image = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRy-Y7t2mawqGyX7E5R54AwWh_0J_i4wNkiYA&usqp=CAU'

icon3 = folium.CustomIcon(
    icon3_image,
    icon_size=(90, 45),
    icon_anchor=(22, 94),
)

folium.Marker(
    location=[40.641766, -73.780968], icon=icon1, popup="JFK Airport"
).add_to(map_hooray)
folium.Marker(
    location=[40.776863, -73.874069], icon=icon2, popup="LGA Airport"
).add_to(map_hooray)
folium.Marker(
    location=[40.689491, -74.174538], icon=icon3, popup="EWR Airport"
).add_to(map_hooray)

Polygon=[[40.773627, -73.993449],[40.765926, -73.999383],[40.751523, -74.009209],[40.699476, -74.019697],[40.710842, -73.976718],[40.729117, -73.970721],[40.742371, -73.972220],[40.758860, -73.958057],[40.768853, -73.981990],[40.773627, -73.993449]]
map_hooray.fit_bounds(Polygon)

def create_convexhull_polygon(
    map_object, list_of_points, layer_name, line_color, fill_color, weight, text
):

    # Since it is pointless to draw a convex hull polygon around less than 3 points check len of input
    if len(list_of_points) < 3:
        return

    # Create the convex hull using scipy.spatial
    form = [list_of_points[i] for i in spatial.ConvexHull(list_of_points).vertices]

    # Create feature group, add the polygon and add the feature group to the map
    fg = folium.FeatureGroup(name=layer_name)
    fg.add_child(
        folium.vector_layers.Polygon(
            locations=form,
            color=line_color,
            fill_color=fill_color,
            weight=weight,
            popup=(folium.Popup(text)),
        )
    )
    map_object.add_child(fg)

    return map_object


# Create a convex hull polygon that contains some points

create_convexhull_polygon(
    map_hooray,
    Polygon,
    layer_name="Example convex hull",
    line_color="#fc0fb1",
    fill_color="#fc0fb1",
    weight=5,
    text="Example convex hull",
)


# Add layer control and show map
folium.LayerControl(collapsed=False).add_to(map_hooray)

# Display the map
map_hooray
```
